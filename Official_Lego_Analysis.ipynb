 How do the number of sets released in 1955 compare to the number of sets released in 2019?

```python
sets_by_year = sets.groupby('year').count()
sets_by_year['set_num'].head()
```

```python
sets_by_year['set_num'].tail(10)
```

```python
plt.plot(sets_by_year.index, sets_by_year.set_num)
```

```python
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
```

```python
themes_by_year = sets.groupby('year').agg({'theme_id': pd.Series.nunique})
```

```python
themes_by_year.rename(columns = {'theme_id':'nr_themes'}, inplace = True)
themes_by_year.head()
```

```python
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2])
```

### Line Charts with Two Seperate Axes

```python
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2])
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
```

```python
ax1 = plt.gca() # get current axes
ax2 = ax1.twinx()

plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2])
```

```python
ax1 = plt.gca() # get current axes
ax2 = ax1.twinx()

#added styling
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2], color='g')
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2], 'b')

ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Sets', color='green')
ax2.set_ylabel('Number of Themes', color='blue')
```

LEGO has licensed many hit franchises from Harry Potter to Marvel Super Heros to many others. But which theme has the largest number of individual sets?

### Merging (i.e., Combining) DataFrames based on a Key


```python
set_theme_count = pd.DataFrame({'id':set_theme_count.index,
                                'set_count':set_theme_count.values})
```

```python
set_theme_count.head()
```

```python
merged_df = pd.merge(set_theme_count, themes, on='id')
merged_df[:3]
```

```python
plt.bar(merged_df.name[:10], merged_df.set_count[:10])
```

```python
plt.figure(figsize=(14,8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Nr of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)

plt.bar(merged_df.name[:10], merged_df.set_count[:10])
```

```python
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown']

plt.figure(figsize=(14, 8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Nr of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)

plt.bar(merged_df.name[:10], merged_df.set_count[:10], color=colors)
plt.show()
```
